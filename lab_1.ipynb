```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,-3.8,1.27,5.38,-2.14])
# a = np.array([1,4.3,0.34,-8.56,2.64])
f = np.poly1d(a)
b = 1+max(abs(a))/abs(a[0])
tochnost = 0.0001
print ('Введённый многочлен:\n', f)
print ('Промежуток, в котором лежат все корни: [',-b,'; ',b,']')
```

```python
x = np.arange(-b, b, 0.1)
plt.plot(x,f(x))
np.set_printoptions(suppress=True)
print (np.column_stack((x, f(x))))
```

```python
r = []
for i in range(len(x)):
    
    if f(x[i-1]) * f(x[i]) < 0:
        
        r.append(x[i-1])
        r.append(x[i])
        
        print (f(x[i-1]),f(x[i]))
        print (f(x[i-1]) * f(x[i]))
        print ('Интервал с корнем: [',x[i-1],';',x[i],']')
        print ()
```

```python
for roots in range(len(r)//2):
    
    start = r[2*roots]
    end = r[2*roots+1]
    print ()
    print ('Для интервала: [',start,';',end,'] находим приблизительное решение:')
    shagi = 0
    
    while (end-start)/2 > tochnost: 

        for i in [start,end]:
            print ('Итерация ',shagi)
            sum = start+end
            c = sum/2
            print (start,c,end)
            if f(c)==0: print ('c= ',c,' точное решение')
            elif f(start)*f(c)<0: end = c
            elif f(start)*f(c)>0: start = c
            shagi = shagi + 1
            
    print ('Заданная точность достигнута: ',(end-start))
    print ('За ',shagi-1,' итерации')
    print ('Приблизительное решение: с=',c)
    print ('Значение функции в данной точке =',f(c))
```
